```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    "student_id": [4, 5, 6, 7, 8],
    "location": ["Sydney", "Tokyo", "Berlin", "London", "Tokyo"],
    "age": [24, 24, 22, 23, 22],
    "sql_marks": [95, 99, 72, 97, 91],
    "excel_marks": [99, 95, 70, 90, 71],
    "python_marks": [55, 89, 99, 74, 79],
    "power_bi_marks": [82, 86, 79, 72, 80],
    "english_marks": [75, 82, 77, 85, 75]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df["age"],df["python_marks"], marker="o" , linestyle="-",color="red") #خطوط مربعة

plt.title("python marks of students")
plt.xlabel("student age")
plt.ylabel("python marks ")
plt.grid(True)
plt.show()
```

```python
plt.bar(df["location"],df["age"] ,color="blue", width=0.3)  #bar=virtical 

plt.title("location vs code")
plt.xlabel("student location")
plt.ylabel("age")
plt.show()
```

```python
plt.barh(df["location"],df["age"] ,color="blue")  #bar=virtical 

plt.title("location vs code")
plt.xlabel("student location")
plt.ylabel("age")
plt.show()
```

```python
plt.plot(df["student_id"],df["python_marks"], marker="*" , linestyle="-",color="red")  #multible line graph with showing the legend 
plt.plot(df["student_id"],df["age"], marker="*" , linestyle="*",color="green")
plt.title("location vs code")
plt.xlabel("student location")
plt.ylabel("age")
plt.legend()
plt.show()
```

```python
df["location"].value_counts().plot.pie(autopct='%1.1f%%', startangle=0)    # Pie chart for student distribution by location
plt.title("Students by Location")
plt.ylabel("")  # Hide default y-label
plt.show()
   
```

```python
plt.hist(df["age"], bins=2, color='skyblue', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.locator_params(axis='y', integer=True) # Set y-axis ticks to integers
plt.show()                                                                 #(no x just y as column) bin= graph shape count
```

```python
plt.scatter(df["python_marks"],df["sql_marks"], marker="^" ,color="red", alpha=0.5 ) #وضوح العلامة كلما كانت اقرب للواحد كانت اوضح
plt.title("python marks vs sql marks")
plt.xlabel("python marks")
plt.ylabel("sql marks ")
plt.show()                                                                                                #scatter plot
```

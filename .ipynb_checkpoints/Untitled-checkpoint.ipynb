```python
import requests
import json
import pandas as pd
```

```python
api_url = "https://fakerapi.it/api/v2/addresses?_quantity=30"
try:
    # 2. Make the GET request to the API
    response = requests.get(api_url)
    
    # 3. Raise an exception for bad status codes (4xx or 5xx)
    response.raise_for_status()

    # 4. Parse the JSON response
    data = response.json()

    print(f"✅ Successfully retrieved {data['total']} addresses.\n")
    
    # 5. Print the first 5 addresses as an example
    print("--- First 5 Retrieved Addresses ---")
    
    # The 'data' field in the response contains the list of addresses
    for i, address in enumerate(data.get('data', [])):
        if i < 5:
            # Use json.dumps for a nicely formatted printout
            print(f"\nAddress {i + 1}:")
            print(json.dumps(address, indent=4))
        else:
            break
            
except requests.exceptions.RequestException as e:
    # Handle any request errors (network, timeout, HTTP errors)
    print(f"❌ An error occurred while fetching data: {e}")
```

```python
address_list = data.get('data', [])
df_addresses = pd.DataFrame(address_list)
df_addresses
```
